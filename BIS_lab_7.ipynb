```python
import numpy as np

def predict_series(prices, steps=50, size=40, window=20):
    x = np.random.choice([-1,0,1], (size,size))
    preds = []
    for t in range(1,len(prices)):
        m = np.tanh((prices[t]-prices[t-1])/prices[t-1])
        for _ in range(steps):
            y = np.roll(x,1,0)+np.roll(x,-1,0)+np.roll(x,1,1)+np.roll(x,-1,1)
            h = m + 0.15*y/4 + 0.1*np.random.randn(size,size)
            x = np.where(h>0.2,1,np.where(h<-0.2,-1,0))
        if t >= window:
            preds.append(np.sign(x.sum()))
        else:
            preds.append(0)
    return np.array(preds)


np.random.seed(0)
prices = np.cumprod(1+0.001*np.random.randn(300))
preds = predict_series(prices)
print(preds)


```
