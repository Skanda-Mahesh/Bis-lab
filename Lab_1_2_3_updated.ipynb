```python
# Partical siumlator

import numpy as np

def predict_series(prices, n=200, steps=10, window=20):
    p = np.random.randn(n,2)
    v = np.random.randn(n,2)*0.1
    preds = []
    for t in range(1,len(prices)):
        m = (prices[t]-prices[t-1])/prices[t-1]
        for _ in range(steps):
            r = p[np.random.randint(0,n)] - p
            d = np.linalg.norm(r,axis=1,keepdims=True)+1e-6
            f = 0.05*r/d
            v = 0.9*v + f + 0.5*m*np.random.randn(n,2)
            p = p + v
        if t >= window:
            preds.append(np.sign(p.sum()))
        else:
            preds.append(0)
    return np.array(preds)

def main():
    np.random.seed(2)
    prices = np.cumprod(1+0.001*np.random.randn(300))
    preds = predict_series(prices)
    print(preds)

if __name__ == "__main__":
    main()

# Gene expression
import numpy as np

def predict_series(prices, steps=40, size=40, window=20):
    a = np.random.randn(size,size)
    b = np.random.randn(size,size)
    c = np.random.randn(size,size)
    preds = []
    for t in range(1,len(prices)):
        m = np.tanh((prices[t]-prices[t-1])/prices[t-1])
        for _ in range(steps):
            e = np.tanh(a*np.sin(b) + c*m + 0.05*np.random.randn(size,size))
            a = np.where(e>0.3,a+0.1*e,a-0.1*e)
            b = np.where(e<-0.3,b+0.1*e,b-0.1*e)
            c = np.tanh(c + 0.1*e)
        if t >= window:
            preds.append(np.sign((a+b+c).sum()))
        else:
            preds.append(0)
    return np.array(preds)

def main():
    np.random.seed(1)
    prices = np.cumprod(1+0.001*np.random.randn(300))
    preds = predict_series(prices)
    print(preds)

if __name__ == "__main__":
    main()

#Gene expression
import numpy as np

def predict_series(prices, steps=40, size=40, window=20):
    g = np.random.randn(size,size)
    h = np.random.randn(size,size)
    preds = []
    for t in range(1,len(prices)):
        m = np.tanh((prices[t]-prices[t-1])/prices[t-1])
        for _ in range(steps):
            e = np.tanh(g*np.sin(h) + m + 0.1*np.random.randn(size,size))
            g = g + 0.2*e
            h = h - 0.2*e
        if t >= window:
            preds.append(np.sign((g+h).sum()))
        else:
            preds.append(0)
    return np.array(preds)

def main():
    np.random.seed(4)
    prices = np.cumprod(1+0.001*np.random.randn(300))
    preds = predict_series(prices)
    print(preds)

if __name__ == "__main__":
    main()


```
